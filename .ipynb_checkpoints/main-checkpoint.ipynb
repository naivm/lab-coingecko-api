##  Imports

```python
import pandas as pd
import requests
import json
```

# Make a dataframe with Bitcoin and Dogecoin market value for last 30 days vs USD

```python
url = 'https://api.coingecko.com/api/v3/coins/dogecoin/market_chart?vs_currency=usd&days=30'
url1 = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=usd&days=30'
response_doge = requests.get(url)#.json()
response_bitcoin= requests.get(url1)
response_bitcoin.content
```

```python
from pandas.io.json import json_normalize
norm_json = json_normalize(dict(response_bitcoin.json()))
ds_normalized = pd.DataFrame(norm_json)

ds_dogecoin = pd.DataFrame(response_doge.json())
ds_bitcoin = pd.DataFrame(response_bitcoin.json())

ds_dogecoin = pd.DataFrame(ds_dogecoin["prices"])
ds_bitcoin = pd.DataFrame(ds_bitcoin["prices"])
#ds_1 = pd.to_datetime(ds_bitcoin['Prices'[0]], unit='ms')
ds_bitcoin.head()
```

```python
ds_doge = pd.DataFrame(ds_dogecoin['prices'].values.tolist(), columns=['Date','Price'])

ds_bit = pd.DataFrame(ds_bitcoin["prices"].values.tolist(), columns=['Date','Price'])
#ds_bitcoin["prices"].apply(pd.Series, index=['Date', 'Prices'])

ds_bit["Date"] = (pd.to_datetime(ds_bit["Date"], unit = "ms"))
ds_doge["Date"] = (pd.to_datetime(ds_doge["Date"], unit = "ms"))

ds_bit.head()
```

```python
ds_bit['Date'] = ds_bit['Date'].dt.date
ds_bit = ds_bit.groupby(['Date'],  as_index=False).mean()

ds_doge['Date'] = ds_doge['Date'].dt.date
ds_doge = ds_doge.groupby(['Date'], as_index=False).mean()
ds_doge.head(10)
```

```python
ds_final = ds_doge.merge(ds_bit, on = "Date")

ds_final= ds_final.rename(columns={"Price_x": "Price Doge", "Price_y": "Price Bit"})
ds_final.head()
```

# Top Ten Coins Report

```python
url2 =  "https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1&sparkline=false"
response_coins = requests.get(url2)
response_coins.headers
```

## List the first 10 coingecko currencies in a DataFrame

```python
ds_coins = pd.DataFrame(response_coins.json())
ds_coins.head()
```

```python
ds_top_ten = ds_coins[['id','name','market_cap','market_cap_change_percentage_24h']].copy()
ds_top_ten.head(10)
```

## Make a csv called 'top_ten_report.csv' and check that it worked
